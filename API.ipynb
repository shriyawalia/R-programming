```python
import requests
import json

session = requests.Session()

query = {
  "query": [
    {
      "code": "Fordonsslag",
      "selection": {
        "filter": "item",
        "values": [
          "PERS"
        ]
      }
    },
    {
      "code": "Bestand",
      "selection": {
        "filter": "item",
        "values": [
          "ITRAF"
        ]
      }
    },
    {
      "code": "Tid",
      "selection": {
        "filter": "item",
        "values": [
          "2004M03",
          "2005M03",
          "2006M03",
          "2007M03",
          "2008M03",
          "2009M03",
          "2010M03",
          "2011M03",
          "2012M03",
          "2013M03",
          "2014M03",
          "2015M03",
          "2016M03",
          "2017M03",
          "2018M03",
          "2019M03",
          "2020M03",
          "2021M03",
          "2022M03",
          "2023M03",
          "2024M03"
        ]
      }
    }
  ],
  "response": {
    "format": "json"
  }
}
```

```python
url = "https://api.scb.se/OV0104/v1/doris/sv/ssd/START/TK/TK1001/TK1001A/Fordon"
```

```python
response = session.post(url, json=query)
response_json = json.loads(response.content.decode('utf-8-sig'))

response_json
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Assuming you have the response JSON stored in response_json
# Extracting relevant data from response_json
data = response_json.get('data', [])
df_data = [{'År': item['key'][2][:4], 'Fordon': int(item['values'][0])} for item in data]

# Creating DataFrame
df = pd.DataFrame(df_data)

# Plotting the data
ax = df.plot(x='År', y='Fordon', kind='line', color='blue', marker='o', linestyle='-')

# Customize the plot
plt.title("Antal fordon enligt bilregistret över tid")
plt.xlabel("År")
plt.ylabel("Antal fordon")
plt.grid(True)

plt.show()
```
